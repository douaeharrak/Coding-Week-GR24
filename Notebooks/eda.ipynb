```python
import pandas as pd
import requests
import zipfile
import io

url = "https://archive.ics.uci.edu/static/public/565/bone+marrow+transplant+children.zip"

# Téléchargement sécurisé
response = requests.get(url)

# Extraction et lecture du CSV spécifique
with zipfile.ZipFile(io.BytesIO(response.content)) as z:
    # Liste les fichiers pour vérifier le nom exact si besoin : print(z.namelist())
    with z.open('bone-marrow.csv') as f:
        df = pd.read_csv(f)

df.head()
```

```python
from ucimlrepo import fetch_ucirepo 

# Récupération du jeu de données via son ID (565 pour Bone Marrow Transplant)
bone_marrow_transplant_children = fetch_ucirepo(id=565) 
  
# Données sous forme de DataFrames Pandas
X = bone_marrow_transplant_children.data.features 
y = bone_marrow_transplant_children.data.targets 

# Fusion pour l'analyse préliminaire (EDA)
df = X.copy()
df['target'] = y

df.head()
```
